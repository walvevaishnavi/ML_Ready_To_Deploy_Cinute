```python
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop("Power Generated",axis=1)
y=data["Power Generated"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression() # library for model
lr=lr.fit(x_train,y_train)#model ko fit karne ke liye
y_pred=lr.predict(x_test)#Prediction model
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print("MAE is",mae)
print("MSE is",mse)
print("R2 is",r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr=dtr.fit(x_train,y_train)

y_pred_dtr=dtr.predict(x_test)
```

```python
mea=mean_absolute_error(y_test,y_pred_dtr)
mse=mean_squared_error(y_test,y_pred_dtr)
r2=r2_score(y_test,y_pred_dtr)
print("MAE is",mae)
print("MSE is",mse)
print("R2 is",r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(n_estimators=100,random_state=42)
rfr=rfr.fit(x_train,y_train)

y_pred_rfr=rfr.predict(x_test)
```

```python
mea=mean_absolute_error(y_test,y_pred_rfr)
mse=mean_squared_error(y_test,y_pred_rfr)
r2=r2_score(y_test,y_pred_rfr)
print("MAE is ",mae)
print("MSE is ",mse)
print("R2 is ",r2)
```

```python
#XGboost
from xgboost import XGBRegressor
xgb=XGBRegressor()
xgb=xgb.fit(x_train,y_train)

y_pred_xgb=xgb.predict(x_test)
```
