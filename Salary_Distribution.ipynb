```python
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
```

```python
from sklearn.model_selection import train_test_split# for spilting the data we import this library
```

```python
x=data.drop("Salary",axis=1)
y=data["Salary"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#0.2 means 20% data will go for testing and 80 % data  for training
#random state 42
#test size and random test nahi likha toh alag data lega, same rakhne ke liye random test and testsizeletehaiwillgo
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression() # library for model
lr=lr.fit(x_train,y_train)#model ko fit karne ke liye
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print("MAE is",mae)
print("MSE is",mse)
print("R2 is",r2)#jtna 1 ke paas hoga utna accha hoga,0 ke paas jitna hoga utna kharab model aur agar 1 hi ho gya toh woh hardcode ho jayega
```

```python
# decision tree
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr=dtr.fit(x_train,y_train)

y_pred_dtr=dtr.predict(x_test)
```

```python
mea=mean_absolute_error(y_test,y_pred_dtr)
mse=mean_squared_error(y_test,y_pred_dtr)
r2=r2_score(y_test,y_pred_dtr)
print("MAE is",mae)
print("MSE is",mse)
print("R2 is",r2)
```

```python
#random forest
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(n_estimators=100,random_state=42)
rfr=rfr.fit(x_train,y_train)

y_pred_rfr=rfr.predict(x_test)
```

```python
mea=mean_absolute_error(y_test,y_pred_rfr)
mse=mean_squared_error(y_test,y_pred_rfr)
r2=r2_score(y_test,y_pred_rfr)
print("MAE is ",mae)
print("MSE is ",mse)
print("R2 is ",r2)
```

```python
#XGboost
from xgboost import XGBRegressor
xgb=XGBRegressor()
xgb=xgb.fit(x_train,y_train)

y_pred_xgb=xgb.predict(x_test)
```
